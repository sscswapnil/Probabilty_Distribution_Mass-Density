# Examples of DATA Distribution

## Uniform Distribution
The probabilities of getting outcome values equally likely thats the basic of Uniform Distribution
Example:- when you roll a fair dice there is equal probability of getting 1 to 6 value

```python
import numpy as np
import matplotlib.pyplot as plt

values=np.random.uniform(-10,10,100000)
plt.hist(values,50)
plt.show()
```

## Normal/Guassian Distribution
This distribution express most of the situation in the universe, thats why its called NORMAL. Mean, Meadian, Mode coincide in this distribution
The curve of distribution is bell shape and symetrical
The total ared under the curve is 1

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

x=np.arange(-3,3,0.001)
plt.plot(x,norm.pdf(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu=5.0
sigma=2.0
values=np.random.normal(mu,sigma,10000)
plt.hist(values,50)
```

## Exponential Distribution function/"Power Law"
Exponential Distribution talks about the interval of timebetween the events 
it is widely used in survival analysis
examples: Time between two metro arrival.
          Time between two calls at call center.
          The life of an Airconditioner.

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

x=np.arange(0,10,0.001)
plt.plot(x,expon.pdf(x))
```

## Poisson Probability mass function
It is applicable only in situations where event occur at random points of time and space where in the interest lies 
only in the number of occurance of that events

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
mu=500
x=np.arange(400,600,0.5)
plt.plot(x,poisson.pmf(x,mu))
```
